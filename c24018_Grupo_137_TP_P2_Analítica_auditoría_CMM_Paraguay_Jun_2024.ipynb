# IMPORTACIÓN DE LAS LIBRERÍAS Y OBTENCIÓN DE DATOS

```python
#   IMPORTACIÓN DE LAS LIBRERÍAS
#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#   IMPORTACIÓN DEL ARCHIVO DE DATOS Y CREACIÓN DE LOS DATAFRAMES
#
ID_planilla = '1BK_7IiD45d0g9xuX2HOUUuxI_gNbk2UP'
url = f'https://docs.google.com/spreadsheets/d/{ID_planilla}/gviz/tq?tqx=out:csv&sheet='

# Sales
df_sales = pd.read_csv(url + 'Sales')


# Distributors
df_distributors = pd.read_csv(url + 'Distributors',thousands=".",decimal=",")


# Exports
df_exports = pd.read_csv(url + 'Exports',thousands=".",decimal=",")

# Locations
df_locations = pd.read_csv(url + 'Locations')
```

# INSPECCIÓN PRELIMINAR, LIMPIEZA Y MODELADO DE DATOS

## SALES

```python
#  INSPECCIÓN PRELIMINAR Y LIMPIEZA DE DATOS
df_sales
```

```python
#  ELIMINACIÓN DE LA ÚLTIMA FILA (FILA DE TOTALES)
df_sales = df_sales.iloc[:-1]
```

```python
# MODELADO DE DATOS: saca el signo $, el '.', y reemplaza ',' por '.' en todas las columnas a partir de Rubber (2da columna), convirtiendo las mismas en float
columnas = df_sales.columns
for n in columnas[range(1,len(df_sales.columns))]:
  df_sales[n] = df_sales[n].str.replace("$","")
  df_sales[n] = df_sales[n].str.replace(".","")
  df_sales[n] = df_sales[n].str.replace(",",".")
  df_sales[n] = df_sales[n].astype(float)



```

```python
# MODELADO DE DATOS: convierte la columna distributor en int
df_sales['distributor'] = df_sales['distributor'].astype(int)
```

```python
df_sales.info()
df_sales.head()
```

```python
# Eliminación de duplicados
df_sales.drop_duplicates(inplace=True)
df_sales.info()
df_sales.head()
```

## DISTRIBUTORS

```python
#  INSPECCIÓN PRELIMINAR Y LIMPIEZA DE DATOS
df_distributors
```

```python
#  ELIMINACIÓN DE FILAS CON VALORES NULOS
df_distributors.dropna(inplace=True)
```

```python
#  MODELADO DE DATOS: elimina en la columna id los '.0', los '.' y la convierte en int. Para ello primero la pasa de float a string
df_distributors['id'] = df_distributors['id'].astype(str)
df_distributors['id'] = df_distributors['id'].str.replace(".0","")
df_distributors['id'] = df_distributors['id'].str.replace(".","")
df_distributors['id'] = df_distributors['id'].astype(int)

df_distributors.info()
df_distributors
```

```python
# Eliminación de duplicados
df_distributors.drop_duplicates(inplace=True)
df_distributors.info()
df_distributors.head()
```

```python
# MODELADO DE DATOS: saca el signo $, el '.', y reemplaza ',' por '.' en todas las columnas a partir de Rubber (2da columna), convirtiendo las mismas en float
columnas = df_exports.columns
for n in columnas[range(1,len(df_exports.columns))]:
  df_exports[n] = df_exports[n].str.replace("$","")
  df_exports[n] = df_exports[n].str.replace(".","")
  df_exports[n] = df_exports[n].str.replace(",",".")
  df_exports[n] = df_exports[n].astype(float)

df_exports.info()
df_exports.head()
```

# ANÁLISIS

## ANÁLISIS DE TOTALES POR PRODUCTO

```python
# EXPORTS: GENERACIÓN DE UN NUEVO DATAFRAME CON LOS TOTALES POR PRODUCTO DE LAS EXPORTACIONES

# Suma de las columnas (todas menos Distributor)
df_tot_exports=df_exports.iloc[:, 1:].sum()

# Arma el DataFrame con los importes en millones
df_tot_exports = pd.DataFrame(df_tot_exports, columns=['Exports'])

# Ordena por importes (columna Exports) de mayor a menor
df_tot_exports.sort_values(by='Exports',ascending=False, inplace=True)

df_tot_exports['Exports']/=1000000
df_tot_exports.rename(columns={'Exports':'Exportaciones (en millones)'}, inplace=True)
df_tot_exports
```

```python
# SALES: GENERACIÓN DE UN NUEVO DATAFRAME CON LOS TOTALES POR PRODUCTO DE LAS VENTAS

# Suma de las columnas (todas menos Distributor)
df_tot_sales=df_sales.iloc[:, 1:].sum()

# Arma el DataFrame con los importes en millones
df_tot_sales = pd.DataFrame(df_tot_sales, columns=['Sales'])

# Ordena por importes (columna Sales) de mayor a menor
df_tot_sales.sort_values(by='Sales',ascending=False, inplace=True)

df_tot_sales['Sales']/=1000000
df_tot_sales.rename(columns={'Sales':'Ventas (en millones)'}, inplace=True)
df_tot_sales
```

```python
# Merge de Total de Exportaciones por Producto vs Total Ventas por Producto
df_Resumen=pd.merge(df_tot_exports, df_tot_sales, left_index=True, right_index=True)
df_Resumen
```

### VISUALIZACIÓN

```python
# Gráfico de barras utilizando df_Resumen
colores=['darkcyan', 'lightsalmon']
df_Resumen.plot(kind='bar', figsize=(9, 6), color=colores)
plt.title('Comparación de Exportaciones y Ventas Totales por Producto')
plt.xlabel('Producto')
plt.ylabel('Importe (en millones)')
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x para mejor legibilidad
plt.tight_layout()  # Ajustar el diseño para evitar recortes
plt.show()
```

En este gráfico se puede observar una diferencia muy marcada entre las exportaciones y ventas de los productos de Steel y Brick dado que las exportaciones en millones son altas pero las ventas estan en niveles nulos. Los otros productos mantienen una regularidad entre los importes de ventas y de las exportaciones.


## ANÁLISIS DE BRICK y DE STEEL POR DISTRIBUIDOR

```python
df_exports.info()
```

```python
# EXPORTS: GENERACIÓN DE UN NUEVO DATAFRAME CON LOS TOTALES POR DISTRIBUIDOR, DE LAS EXPORTACIONES DE STEEL Y BRICK

# Selecciona las columnas a analizar
df_dist_exports=df_exports.loc[:, ['distributor', 'Brick', 'Steel']]

# Divide importes en millón para presentar y analizar
df_dist_exports['Brick']/=1000000
df_dist_exports['Steel']/=1000000

df_dist_exports
```

```python
# SALES: GENERACIÓN DE UN NUEVO DATAFRAME CON LOS TOTALES POR DISTRIBUIDOR, DE LAS VENTAS DE STEEL Y BRICK

# Selecciona las columnas a analizar
df_dist_sales=df_sales.loc[:, ['distributor', 'Brick', 'Steel']]

# Divide importes en millón para presentar y analizar
df_dist_sales['Brick']/=1000000
df_dist_sales['Steel']/=1000000

df_dist_sales
```

```python
# Merge de las Exportaciones por Distribuidor vs Total Ventas por Distribuidor
df_Resumen_dist=pd.merge(df_dist_exports, df_dist_sales, on='distributor')
df_Resumen_dist
```

```python
# Cambia el orden de las columnas y las renombra
df_Resumen_dist_2 = df_Resumen_dist.reindex(['distributor',	'Brick_x', 'Brick_y',	'Steel_x', 'Steel_y'], axis=1)
df_Resumen_dist_2.rename(columns={'Brick_x':'Brick Exportaciones', 'Brick_y':'Brick Ventas', 'Steel_x':'Steel Exportaciones', 'Steel_y':'Steel Ventas'}, inplace=True)
df_Resumen_dist_2
```

### VISUALIZACIÓN

```python
# prompt: qué gráfico puedo hacer con df_Resumen_dist_2

# Gráfico de barras agrupadas para comparar exportaciones y ventas de Brick y Steel por distribuidor
df_Resumen_dist_2.plot(x='distributor', y=['Brick Exportaciones', 'Brick Ventas', 'Steel Exportaciones',
                                           'Steel Ventas'], kind='bar', figsize=(15, 6), width=1.0)
plt.title('Comparación de Exportaciones y Ventas de Brick y Steel por Distribuidor')
plt.xlabel('Distribuidor',fontsize=14)
plt.ylabel('Importe (en millones)',fontsize=14)
plt.yscale('log')
plt.yticks(fontsize=12)
plt.xticks(rotation=90, ha='right', fontsize=12)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
print()
print()

# Gráfico de dispersión para analizar la relación entre exportaciones y ventas de Brick
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Brick Exportaciones', y='Brick Ventas', data=df_Resumen_dist_2)
plt.title('Relación entre Exportaciones y Ventas de Brick')
plt.xlabel('Exportaciones de Brick (en millones)')
plt.ylabel('Ventas de Brick (en millones)')
#plt.xscale('linear')
plt.grid(True)
plt.show()
print()
print()

# Gráfico de dispersión para analizar la relación entre exportaciones y ventas de Steel
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Steel Exportaciones', y='Steel Ventas', data=df_Resumen_dist_2)
plt.title('Relación entre Exportaciones y Ventas de Steel')
plt.xlabel('Exportaciones de Steel (en millones)')
plt.ylabel('Ventas de Steel (en millones)')
plt.grid(True)
plt.show()
print()
print()

# Gráfico de caja para comparar la distribución de exportaciones y ventas de Brick y Steel
df_Resumen_dist_2_melted = pd.melt(df_Resumen_dist_2, id_vars='distributor', value_vars=['Brick Exportaciones', 'Brick Ventas', 'Steel Exportaciones', 'Steel Ventas'])
plt.figure(figsize=(10, 6))
sns.boxplot(x='variable', y='value', data=df_Resumen_dist_2_melted)
plt.title('Distribución de Exportaciones y Ventas de Brick y Steel')
plt.xlabel('Tipo de Transacción')
plt.ylabel('Importe (en millones)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

En el gráfico de comparación de exportaciones y ventas de Brick y Steel por Distribuidor se puede divisar que el distribuidor 1183 contiene exportaciones de Steel elevadas en relación a sus pares. En cuanto a las exportaciones de Brick los distribuidores 1463, 1560 y 815 son los distribuidores que se destacan por encima del resto sin embargo, para las ventas de Brick y Steel, en
general los distribuidores mantienen importes en millones bajos con respecto a
las exportaciones.
